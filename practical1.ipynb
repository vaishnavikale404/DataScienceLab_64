WAP to compute factorial of a given number

```python
a= int(input("enter the first number:"))
fact=1
i=1
while i<=a:
    fact=fact*i
    i=i+1
print("factorial of",a,"is",fact)
```

WAP to display table for a given number



```python
def multiplication(n,limit=10):
   print(f"multiplication table of {n}")
   for i in range(1,limit+1):
     print(f"{n}x{i}={n*i}")
input_number = 5
multiplication(input_number)
print("\n")
```

WAP to identify the given number is odd or even

```python
a = int(input("enter the number:"))
def even_odd(n):
  if n%2==0:
     return f"{n} is even number"
  else:
    return f"{n} is an odd number"
print(even_odd(n=a))
```

WAP to identify the entered number is palindrome or not.



```python

number = int(input("Enter a number to check if it's a palindrome: "))

# Convert the number to a string
number_str = str(number)

# Reverse the string
# The [::-1] slice creates a reversed copy of the string
reversed_str = number_str[::-1]

# Compare the original string with the reversed string
if number_str == reversed_str:
    print(f"{number} is a palindrome!")
else:
    print(f"{number} is NOT a palindrome.")
```

Develop an Expert System that provides simple decision-making.

### Simple Decision-Making Expert System (Movie Recommender)

```python
def movie_recommender():
    print("Hello! I'm here to recommend a movie based on your mood.")
    print("How are you feeling today? (e.g., happy, sad, bored, excited)")

    # Get user's mood as input
    mood = input("Enter your mood: ").lower() # .lower() converts input to lowercase for easier comparison

    # Decision-making rules
    if "happy" in mood or "excited" in mood:
        recommendation = "a Comedy or an Adventure movie!"
    elif "sad" in mood or "gloomy" in mood:
        recommendation = "an Uplifting Drama or a feel-good Romantic Comedy."
    elif "bored" in mood or "nothing to do" in mood:
        recommendation = "a Thriller or a Sci-Fi movie to get your mind working!"
    elif "tired" in mood or "relaxed" in mood:
        recommendation = "a Documentary or a Light Drama."
    else:
        recommendation = "I'm not sure what to recommend for that mood. How about a Classic?"

    print(f"\nBased on your mood, I recommend you watch {recommendation}")

# Call the expert system
movie_recommender()
```
